```python
SQL_TIPOSPROCESSO = 'SELECT * FROM ovr_tiposprocesso'
SQL_PROCESSOS_OVR = 'SELECT o.id, t.descricao, p.numero FROM ovr_ovrs o ' + \
                    'inner join ovr_processos p on o.id = p.ovr_id ' + \
                    'inner join ovr_tiposprocesso t on t.id = p.tipoprocesso_id '
SQL_OVR = 'SELECT * FROM ovr_ovrs'

def get_df_tiposprocesso(engine, sql_tiposprocesso=SQL_TIPOSPROCESSO):
    return pd.read_sql(sql_tiposprocesso, engine)

def get_df_processos_ovr(engine, sql_processos_ovr=SQL_PROCESSOS_OVR):
    df = pd.read_sql(sql_processos_ovr, engine)
    df = df.pivot(index='id', columns='descricao', values='numero')
    df = df.fillna('')
    df = df.reset_index()
    return df

def get_df_ovr(engine, sql_ovr=SQL_OVR):
    df = pd.read_sql(sql_ovr, engine)
    return df

def get_df_controle(df_ovr, df_processos):
    return pd.merge(df_ovr, df_processos, how='inner', on='id')
```

```python
from bhadrasana.models.ovr import OVR, ProcessoOVR, EventoOVR, TGOVR
session.rollback()
```

```python
def _decimal(texto):
    novotexto = texto.replace('.', '').replace(',', '.')
    try:
        return float(novotexto)
    except:
        return 0.
    
def monta_ovr(registro):
    try:
        ovr = OVR()
        ovr.numeroCEmercante = registro['CE'].strip().split()[0]
        ovr.observacoes = registro['Apelido'].strip()
        recintostr = str(registro['RECINTO']).lower().split()[0]
        ovr.recinto_id = get_recinto(recintostr)
        ovr.datahora = datetime.strptime(registro['datahora'], '%y/%m/%d')
        ovr.ano = str(ovr.datahora.year)
        ovr.numero = registro['ABERTURA']
        ovr.fase = 4
        ovr.tipooperacao = 1
        ovr.tipoevento_id = 14
        #print('OVR', ovr.__dict__)
        session.add(ovr)
        session.flush()
        session.refresh(ovr)
        #print('OVR', ovr.__dict__)
    except Exception as err:
        print(err)
        return 
    try:
        tg = TGOVR()
        tg.ovr = ovr
        tg.descricao = registro['MERCADORIA']
        tg.valor = _decimal(registro['VALOR R$'])
        tg.peso = _decimal(registro['PESO'])
        session.add(tg)
        #print('TG', tg.__dict__)
        for tipo, numero in [(6, 'FICHA RADAR'),
                            (1, 'PAF'),
                            (4, 'PAF RFFP')]:
            if not isinstance(numero, str):
                continue
            processo = ProcessoOVR()
            processo.ovr = ovr
            processo.numero = registro[numero]
            processo.tipoprocesso_id = tipo
            session.add(processo)
            #print('Processo', processo.__dict__)
        session.commit()
    except Exception as err:
        session.rollback()
        print(err)
    #print(registro)
    

    
```
