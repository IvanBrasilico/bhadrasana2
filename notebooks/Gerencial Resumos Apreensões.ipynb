```python
# SQL para puxar todos os dados de apreensões de cocaína
SQL_APREENSOES = '''
SELECT ovr.id, ovr.datahora, ovr.tipooperacao, count(a.id) as Qtde, sum(a.peso) as Peso
  FROM ovr_ovrs ovr
 inner join ovr_verificacoesfisicas rvf on rvf.ovr_id = ovr.id
 inner join ovr_apreensoes_rvf a on a.rvf_id = rvf.id
 inner join ovr_tiposapreensao t on t.id = a.tipo_id
 where a.tipo_id=1
 group by ovr.id'''
```

```python
SQL = SQL_APREENSOES
df = pd.read_sql(SQL, engine)
df.head()
```
